```python
import pandas as pd
df = pd.read_csv(r"C:\Users\91952\Downloads\train.csv")
```

```python
df.info()
df.describe()
df.value_counts()
df.head()
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt

plt.hist(df['Age'], bins=30, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
survival_counts = df['Survived'].value_counts()
plt.bar(survival_counts.index, survival_counts.values, color=['red', 'green'])
plt.title('Survival Counts')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
plt.boxplot(df['Fare'])
plt.title('Fare Distribution')
plt.ylabel('Fare')
plt.show()
```

```python
gender_survival = df.groupby('Sex')['Survived'].mean()
plt.bar(gender_survival.index, gender_survival.values, color=['blue', 'pink'])
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.show()
```

```python
pclass_survival = df.groupby('Pclass')['Survived'].mean()
plt.bar(pclass_survival.index, pclass_survival.values, color='purple')
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
plt.scatter(df['Age'], df['Fare'], alpha=0.5)
plt.title('Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```
